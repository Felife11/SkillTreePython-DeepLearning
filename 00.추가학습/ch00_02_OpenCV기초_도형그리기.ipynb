## 이미지 준비
---

````python
path = 'sea.png'
img = cv2.imread(path)
print(img.shape)
h, w, c = img.shape
print('* 이미지 높이 : ', h)
print('* 이미지 너비 : ', w)
print('* 채널 수 : ', c)
cv2_imshow(img)
````

### **| 색깔 파라미터 설정**

````python
B = 255 #@param {type: "slider", min : 0, max : 255}
G = 0 #@param {type: "slider", min : 0, max : 255}
R =244 #@param {type: "slider", min : 0, max : 255}
shape = (100,100, 3)
value = (B, G,R)  # B, G, R
도화지 = np.full(shape, value, np.uint8) #shape 크기의 value 생성
#print(도화지)
cv2_imshow(도화지)
````

## 이미지 크롭
---

### **| 슬라이싱을 이용한 이미지 크롭**

````python
#슬라이싱
croped = img[0:100, 0:100] # 세로를 0~100 픽셀만 뽑고, 가로도 0~100 픽셀만 뽑느다.
cv2_imshow(croped)
````

#### **크롭 파라미터 설정**

````python
X_start = 156 #@param {type: "slider", min : 0, max : 635, step : 1}
X_end = 333 #@param {type: "slider", min : 0, max : 635,step : 1}
Y_start =59 #@param {type: "slider", min : 0, max : 382, step : 1}
Y_end =224 #@param {type: "slider", min : 0, max : 382, step : 1}
crp = img[Y_start:Y_end, X_start:X_end] # 세로를 0~100 픽셀만 뽑고, 가로도 0~100 픽셀만 뽑느다.
cv2_imshow(crp)
````

## 도형 그리기
---

### **| Line 그리기**

Start와 End 점을 연결하여 직선을 그립니다.

#### `cv2.line(img, start, end, color, thickness)`
* Parameters
  * img – 그림을 그릴 이미지 파일
  * start – 시작 좌표(ex; (0,0))
  * end – 종료 좌표(ex; (500. 500))
  * color – BGR형태의 Color(ex; (255, 0, 0) -> Blue)
  * thickness (int) – 선의 두께. pixel

````python
cv2.line(img, (X_start, Y_start), (X_end, Y_end), value, 10)
cv2_imshow(img)
````

### **| 사각형 그리기**

top-left corner와 bottom-right corner점을 연결하는 사각형을 그립니다.

#### `cv2.rectangle(img, start, end, color, thickness)`
* Parameters
  * img – 그림을 그릴 이미지
  * start – 시작 좌표(ex; (0,0))
  * end – 종료 좌표(ex; (500. 500))
  * color – BGR형태의 Color(ex; (255, 0, 0) -> Blue)
  * thickness (int) – 선의 두께. pixel

* Sample Code
```
img = cv2.rectangle(img, (384, 0), (510, 128), (0,255,0), 3)
```

````python
cv2.rectangle(img, (X_start, Y_start), (X_end, Y_end), value, 5)
cv2_imshow(img)
````

### **| 원 그리기**

top-left corner와 bottom-right corner점을 연결하는 원을 그립니다.


* Parameters
  * img – 그림을 그릴 이미지
  * center – 원의 중심 좌표(x, y)
  * radian – 반지름
  * color – BGR형태의 Color
  * thickness – 선의 두께, -1 이면 원 안쪽을 채움

* Sample Code
```
img = cv2.circle(img, (447,63), 63, (0,0,255), -1)
```

````python
cv2.circle(img, (X_start, Y_start), 50, value, 5)
cv2_imshow(img)
````

### **| 텍스트 추가**

#### `cv2.putText(img, text, org, font, fontSacle, color, thickness)`
* Parameters
 * img – image
 * text – 표시할 문자열
 * org – 문자열이 표시될 위치. 문자열의 bottom-left corner점
 * font – font type. CV2.FONT_XXX
 * fontSacle – Font Size
 * color – fond color
 * thickness - 선의 두께

* Sample Code
```
cv2.putText(img, 'OpenCV', (10,500), cv2.FONT_HERSHEY_SIMPLEX, 4, (255,255,255), 2)
```

````python
cv2.putText(img, 'Hi', (X_start, Y_start), cv2.FONT_HERSHEY_DUPLEX, 2, value, 2)
cv2_imshow(img)
````
